```python
%load_ext autoreload
%autoreload 2

import nest
import sys
import os.path
import numpy as np
from math import degrees
import matplotlib.pyplot as plt
from itertools import groupby

sys.path.append('../')
import params
import nest_tools
import mnist_tools
```

```python
def load_matrices(file, cores=1):
    matrix = np.zeros((params.NE,params.NE))
        
    for i in range(cores):
        filename = file+'.'+str(i)
        data = np.load(filename+'.npy')
        
        matrix += data
    return matrix

final_connectivity = load_matrices("data/final_connectivity", cores=4)
```

```python
# reconstruct final connectivity from changes
init = np.zeros((4000, 4000))
for t in [0,1,2,3]:
    for p in [0,1,2,3]:
        m = np.load("data/matrix_change"+str(t)+"."+str(p)+".npy")

        init += m.item().toarray()
        
```

```python
np.all(final_connectivity == init)
```
